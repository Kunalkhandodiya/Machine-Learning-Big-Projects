```python
# data Represent the Bar plot
plt.figure(figsize=(8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys())
        ,list(ipl['player_of_match'].value_counts()[0:5]),color='g')

plt.show()
```

```python
# represent the match results 
ipl['result'].value_counts()
```

```python
# represent the toss results
ipl['toss_winner'].value_counts()
```

```python
#Extracting the Records where a team won batting First
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
#looking by the head
batting_first.head()
```

```python
# making a histogram
plt.figure(figsize=(5,7))
plt.hist(batting_first['win_by_runs'])
plt.title('Distribution the Runs')
plt.xlabel('Runs')
plt.show()
```

```python
# Finding out the number of wins w.r.t. each team after batting first
batting_first['winner'].value_counts()
```

```python
# represent the batting-First data in barplots in below
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color='g')
plt.show()
```

```python
#plt.figure(figsize=(7,7))
```

```python
# Makin a Pie Chart on batting_first data
plt.figure(figsize=(10,10))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#extracting those records where a team has won after batting second
batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
# Looking at the head
batting_second.head()
```

```python
# making a histogram for frequency of wins w.r.t. number of wickets
plt.figure(figsize=(5,5))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.grid(True)
plt.show()
```

```python
# finding out the frequency of numbers of win w.r.t each time after batting second
batting_second['winner'].value_counts()
```

```python
#making a bar plot for top3 team with most wins after batting second

plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color=["y","b","g"])
plt.show()
```

```python
# making a pie chart for distribution of most wins after batting second

plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
